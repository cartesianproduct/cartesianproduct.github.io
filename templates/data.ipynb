```python
from collections import OrderedDict
d = {}
for i in range(len(base)):
    s = {base['Ward/VillageTract'][i]:{'tot_pop':base['tot_pop'][i],
                                   'male_pop':base['male_pop'][i],
                                   'female_pop':base['female_pop'][i]}}
    #s = OrderedDict(sorted(d.items(), key=lambda t: t[0]),reverse=True)
    d.setdefault(base['Township'][i],[]).append(s)
        
```

```python
d
```

```python
b = pd.read_csv('Base Township Data  - township_data_refined_new.csv')
```

```python
b['Ward/Village'] = [d[i] if i in d.keys() else 0 for i in b['township_name']]
```

```python
b.to_csv('Base Township Data  - township_data_refined_new.csv', index = False)
```
